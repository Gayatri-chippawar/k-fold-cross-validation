```python
#importing depenedencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dataset=pd.read_csv('datasets/heart_disease_data.csv')
```

```python
dataset.head()
```

```python
x=dataset.drop(columns='target',axis=1)
y=dataset.target
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
```

### comparing the models


```python
models=[LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
def test_train_models():
    for i in models:
        i.fit(x_train,y_train)
        test_prediction=i.predict(x_test)
        acc=accuracy_score(test_prediction,y_test)
        print(f"the accuracy score of {i}:{acc}")
```

```python
test_train_models()
```

### Cross Validation

```python
models=[LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
val_lr=cross_val_score(LogisticRegression(max_iter=1000),x,y,cv=5)
print(val_lr)
mean_accuracy=sum(val_lr)/len(val_lr)
print(mean_accuracy)
```

```python
def cross_validation():
    for model in models:
        cv_score=cross_val_score(model,x,y,cv=5)
        print(cv_score)
        mean_accuracy=sum(cv_score)/len(cv_score)
        print(f"The mean accuracy score of {model} is :{mean_accuracy*100:%,.2f}")
```
